```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("boston_house_prices.csv", header=1)
df
```

```python
df[df.duplicated()] # TO check whether the dataset is having duplicates or not
df.isnull().sum() # To check the null values
```

```python
# Independent Features and Dependent Features
x = df.iloc[:, :13]
y = df.MEDV
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test = train_test_split( x , y, test_size=0.30 , random_state=42)
```

```python
x_train
```

```python
# Standardizing the dataset
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test= scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
reg_predict = regression.predict(x_test)
```

```python
sns.displot(reg_predict - y_test,kind='kde') # kde is used for the better visualization of the prediction and the real data
```

```python
from sklearn.metrics import r2_score
score = r2_score(reg_predict,y_test)
score
```
